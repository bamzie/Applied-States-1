**A.1(c) [10 points] Three dice are thrown. What is the probability that a sum of 12 appears on the faces? What is the probability that a sum of 13 appears?**

To find the probability that a sum of 12 appears on the faces of 3 dices, we must first look at the possible sums from 3 dices. 

$D_{12} = \{(6,5,1),(4,3,5),(5,5,2),(6,4,2),(6,3,3),(4,4,4)\}$

Using permutations algorithm, $P(n,r) = \frac{n!}{(n-r)!}$ we find that for elements in the set that have 3 different 
values the combination is 6, for elements that have 2 different values the combination is 3, and for 1 distinct value 
the combinatin is 1. 

$(6,5,1) = P(3,2) = \frac{3!}{(3-2)!} = 6$, 
$\newline(4,3,5) = P(3,2) = \frac{3!}{(3-2)!} = 6$
$\newline(5,5,2) = P(3,1) = \frac{3!}{(3-1)!} = 3$
$\newline(6,4,2) = P(3,2) = \frac{3!}{(3-2)!} = 6$
$\newline(6,3,3) = P(3,1) = \frac{3!}{(3-1)!} = 3$
$\newline(4,4,4) = P(3,0) = \frac{3!}{(3-0)!} = 1$

Therefore, their are 25 different combinations. 
$\newline P(D_{12}) = \frac{25}{6^3} \approx 0.1157$

We do similar work for the probability that a sum of 13 appears on the faces of 3 dices. 

$P(D_{13}) = \{(5,4,4),(4,6,3),(6,6,1), (5,5,3),(4,3,6) \}$

$P(D_{13}) = \frac{21}{216} = \approx 0.097$

## A.2 Probability and causality!

What does it mean for one event C to cause another event E—for example, smoking (C) to cause cancer (E)? There is a long history in philosophy, statistics, and the sciences of trying to clearly analyze the concept of a cause. One tradition says that causes raise the probability of their effects; we may write this symbolically is
$$
\begin{equation} 
P(E | C)>P(E). 
\end{equation} \qquad \qquad (1)
$$ 


**A.2(a) [6 points] Does equation (1) imply that $P(C | E) > P(C)$? If so, prove it. If not, give a counter example.**




No. From looking at the example given above, smoking (C) to cause cancer (E), the opposite is not true. 
Cancer(E) to cause smoke (C) is no true. Therefore $P(C|E) > P(C)$ in not true in this case. 

Another way to formulate a probabilistic theory of causation is to say that
$$P (E | C) > P(E | C^C). \qquad \qquad (2)$$ 

**A.2(b) [6 points] Show that equation (1) from A.2(a) implies equation (2).**

We want to show that $P(E|C) > P(E) \Rightarrow P(E|C) > P(E|C^c$)
$\newline$First, lets note that $P(E) = P(E|C)P(C) + P(E|C^c)P(C^c)$ and $P(C^c) = 1 - P(C)$

$P(E|C) > P(E)$
$\newline \Rightarrow P(E|C) > P(E|C)P(C) + P(E|C^c)P(C^c)$ 
$\newline \Rightarrow P(E|C) - P(E|C)P(C) > P(E|C^c)(1-P(C))$
$\newline \Rightarrow P(E|C)(1 - P(C)) > P(E|C^c)(1-P(C))$

Divide $1-P(C)$ from both sides and we obtain: 

$P(E|C) > P(E|C^c) $

## A.3 Conditional probabilities

Suppose a particular crime is committed in Jerry’s apartment. We’d like to know whether Newman is guilty of the crime. We are torn as to whether we think he is guilty: we think it’s equally likely that he guilty or not guilty. Suppose that, in similar situations, we know that if a suspect is guilty, 85% of the time their finger prints are found at the scene, and, we know that if a suspect is not guilty, 30% of the time their finger prints are found at the scene.

**A.3(a) [5 points] What is the probability that Newman’s finger prints are found at the scene?**

Let $G_1 =$ suspect guilty, $G_2 = $ suspect not guilty, and $F = $ finger print found at scene

$P(F|G_1) = 0.85$ and $P(F|G_2) = 0.30$ and $P(G_1) = 0.50$ and $P(G_2) = 0.50$

From LOTP, we can utilize $P(F) = P(F|G_1)P(G) + P(F|G_2)P(G_2)$

$P(F) = 0.575$

**A.3(b) [5 points] If Newman’s finger prints are found at the scene, how likely is it that he is guilty?**

$P(G_1|F) = \frac{P(F|G_1)P(G_1)}{P(F)} = \frac{0.85(0.5)}{0.575} = 0.739$

**A.3(c) [7 points] If Newman’s finger prints are not found at the scene, how likely is it that he is guilty?**

$P(G_1|F^c)= \frac{P(F^c|G_1)P(G_1)}{P(F^c)} = \frac{0.15(0.5)}{0.425} = 0.174$

## B.1 Probability Simulations

There is an urn that has 100 marbles in it: 2 are red, 70 are yellow, 13 are green and 15 are blue. Simulate data to imitate someone drawing a marble from the urn (replacing after each draw), with the color recorded at each draw (HINT: the ${\tt sample()}$ function in ${\tt R}$ will be useful here).

**B.1(a) [8 points] Plot the fraction of times a yellow marble is drawn in the first 1000 draws. On the x-axis, include the number of draws, and on the y-axis, include the cumulative fraction of yellow draws. Why does this number fluctuate more when the number of draws is smaller?**

```r
#YOUR CODE HERE
set.seed(001);
#red = r, yellow = y, green = g, blue = b
urn <- c(rep('r',2), rep('y',70), rep('g',13), rep('b',15))
n = 1000
s = sample(urn, size = 1, replace = TRUE);  
vec_fract = matrix(NA, ncol = 1, nrow=n)
y = 0
for (i in 1:n){
    x = sample(urn,i, replace = TRUE)
    for(j in 1:length(x)){
        y = y + ifelse(x[j] == 'y', 1, 0)
    }
    y = y/length(x)
    vec_fract[i] = y;
}
plot(vec_fract, type ="l", ylab = "Fraction of Yellow", xlab = "Draws", col = "gold" )
abline(0.7, 0, col='red')
```

The number fluctuates more when the numbers of draw is smaller because the smaller the draw the less data there is to calculate the fraction of draws properly. When there are more draws the we can approximate better since there are more outcomes to calculate.
