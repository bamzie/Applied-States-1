**A.1(a) [8 points] The CU Boulder triathlon team has 12 women and 9 men. The team is going to a race and can only enter 5 participants. What is the probability of randomly selecting a race squad of 5 participants with exactly 3 women?**

Number of ways to choose 3 women from 12 total:
$$ C_{12,3} = \binom{12}{3} $$

Number of ways to choose 2 men from 9 total:
$$ C_{9,2} = \binom{9}{2} $$

Number of total possible combinations for selecting 5 people from a group of 21:
$$ C_{21,5} = \binom{21}{5} $$

Thus, we find the probability of selecting a team of 5 with exactly 3 women to be:

$$ \frac{C_{12,3}*C_{9,2}}{C_{21,5}} = \frac{\binom{12}{3}*\binom{9}{2}}{\binom{21}{5}} = \frac{220*36}{20349} \approx 0.389$$

Alternatively,
$$ P(3 Women) = \left(\frac{12}{21}\right)*\left(\frac{11}{20}\right)*\left(\frac{10}{19}\right)*\left(\frac{9}{18}\right)*\left(\frac{8}{17}\right)*C_{5,3}=\frac{95040}{2441880}*\binom{5}{3} \approx 0.389 $$

## A.2 Probability and causality!

What does it mean for one event C to cause another event E—for example, smoking (C) to cause cancer (E)? There is a long history in philosophy, statistics, and the sciences of trying to clearly analyze the concept of a cause. One tradition says that causes raise the probability of their effects; we may write this symbolically is
$$
\begin{equation} 
P(E | C)>P(E). 
\end{equation} \qquad \qquad (1)
$$ 


**A.2(a) [6 points] Does equation (1) imply that $P(C | E) > P(C)$? If so, prove it. If not, give a counter example.**




Here we apply Bayes' Rule:


\begin{align*}
P(C\vert E) & =\frac{P(E\vert C)P(C)}{P(E)}\\
 & >\frac{P(E)P(C)}{P(E)}\\
 & =P(C)
\end{align*}

Thus, if we subscribe to the above definition of causality, cancer causes smoking!

Another way to formulate a probabilistic theory of causation is to say that
$$P (E | C) > P(E | C^C). \qquad \qquad (2)$$ 

**A.2(b) [6 points] Show that equation (1) from A.2(a) implies equation (2).**

\begin{align*}

\implies P(E\vert C)P(C^{C}) & >P(E|C^{C})P(C^{C})\\
\implies P(E\vert C) & >P(E|C^{C})

## A.3 Conditional probabilities

Suppose a particular crime is committed in Jerry’s apartment. We’d like to know whether Newman is guilty of the crime. We are torn as to whether we think he is guilty: we think it’s equally likely that he guilty or not guilty. Suppose that, in similar situations, we know that if a suspect is guilty, 85% of the time their finger prints are found at the scene, and, we know that if a suspect is not guilty, 30% of the time their finger prints are found at the scene.

**A.3(a) [5 points] What is the probability that Newman’s finger prints are found at the scene?**

We are given the following probabilities,

Probability that Newman is guilty: $$P(G)=0.5$$

Probability the Newman is not guilty: $$P(-G)=0.5$$

Define the probability that finger prints are found as: 

$$P(F)$$

Probability that a suspect's finger prints are found at the scene given the suspect is guilty: 

$$P(F|G)=0.85$$

Probability that a suspect's finger prints are found at the scene given the suspect is not guilty: 

$$P(F|-G)=0.30$$


$ P(G)$ and $P(-G)$ partition the sample space as Newman is either guilty or not guilty and $P(G)+P(-G)=0.5+0.5=1$ so we can use the Law of Total Probability.

$$ P(F) = P(F|G)P(G)+P(F|-G)P(-G) $$

$$= (0.85)(0.5)+(0.30)(0.5) = 0.575 $$

Thus, the probability that Newman's finger prints are found at the scene is 0.575 or 57.5%.

**A.3(b) [5 points] If Newman’s finger prints are found at the scene, how likely is it that he is guilty?**

This problem is asking us for the probability that Newman is guilty, given that his finger prints are found at the scene. i.e. $$P(G|F)$$
We find $P(G|F)$ using Bayes' Theorem:
$$P(G|F)=\frac{P(F|G)P(G)}{P(F)}$$

$$P(G|F)= \frac{(0.85)(0.5)}{(0.575)} = 0.739$$

Thus, the probability that Newman is guilty given his fingerprints were found at the crime scene is $P(G|F)= 0.739.$

**A.3(c) [7 points] If Newman’s finger prints are not found at the scene, how likely is it that he is guilty?**

This problem is asking us to find the probability that Newman is guilty given that his finger prints are not found at the scene i.e.
$$P(G|-F)$$
Define $P(-F)$ as the probability that Newman's finger prints are not found.

We know from part(a) that the probability that Newton's finger prints are found at the scene is $$P(F)=0.575$$

Further, $P(F)$ and $P(-F)$ partition the sample space as Newton's finger prints are either found at the scene or not found at the scene. And,
$$P(-F)=1-P(F)=1-0.575=0.425$$

From the Law of Total Probability, we find:

$$P(G)= P(G|-F)P(-F)+P(G|F)P(F)$$

$$ P(G|-F)= \frac{P(G)-P(G|F)P(F)}{P(-F)} $$

$$ P(G|-F)= \frac{(0.5)-(0.739)(0.575)}{(0.425)} = 0.177$$

Thus, the probability that Newman is found guilty given that his finger prints were not found at the scene is 0.177.

## A.4 More practice (not graded!) 

The game of Yahtzee is played with five fair dice. The goal is to roll certain ‘hands’, such as Yahtzee (all five dice showing the same number) or a full house (three of a kind and two of a kind). In the first round of a player’s turn, the player rolls all five dice. Based on the outcome of that roll, the player has a second and third round, where he/she can then choose to re-roll any subset of the dice to get a desired hand.

**A.4(a) What is the probability of rolling a Yahtzee on the first round?**

There are 6 winning outcomes (all ones, or all twos, etc...) from among $6^5$ outcomes. Thus the probability of rolling a Yahtzee in round one is 

$$\frac{6}{6^5}=6^{-4}$$

**A.4(b) Suppose that, on the second round, the dice are {2, 3, 4, 6, 6}. You decide to re-roll both sixes in the third round. What is the probability that you roll either a small straight or a large straight (a large straight is where all five dice are in a row)?**

If we decide to reroll the two 6’s, then we start with the 3 dice combination of (2,3,4). For the last two dice, we can role either (5,2), (5,3),...,(5,6), (2,5), (3,5),..., (6,5), (1,1),...,(1,4),(1,6), (1,1),...,(4,1), or (6,1).
So, we can see that the probability of getting either a small or a large straight is:
$2(1/6)(5/6) + 2(1/6)(5/6) = 5/9 \approx 0.556.$

**A.4(c) A small straight is defined as having 4 of the 5 dice all in a row (for example, {1, 2, 3, 4, 6}). What is the probability of rolling a small straight on the first round?**

Note that order in which the dice land does not matter.

There are $6^5$ possible outcomes, among which we have the following possible small straights:

 1. {1,2,3,4,x} where x≠5
 2. {2,3,4,5,x} where x≠6, x≠1
 3. {3,4,5,6,x} where x≠2
 
However, it's best to think of this in terms of outcomes with all unique numbers versus those with repeated numbers:

Unique:
 1. {1,2,3,4,x} where x=6
 2. {3,4,5,6,x} where x=1

Repeated:
 1. {1,2,3,4,x} where x is 1, 2, 3, or 4
 2. {2,3,4,5,x} where x is 2, 3, 4, or 5
 3. {3,4,5,6,x} where x is 3, 4, 5, or 6

For each of the unique numbers outcomes, there are $4!$ orderings of the numbers in the straight, and $5\choose1$ places to put the remaining element (x).

For each of the repeated numbers outcomes, it's more complicated as two of the numbers are exchangeable. That is, e.g., {1,2,3,4,x} = {x,2,3,4,1} when x = 1. We have to be careful to not double count these. So, for each possible value of x, there are $3!$ orderings of remaining unique numbers and $5\choose2$ places to put the repeated numbers, yielding a total number of $4\cdot 3! {5\choose2}$ possibilities for each outcome.

The total probability is then $$\frac{2\left(4! {5\choose1}\right) + 3\left(4\cdot 3! {5\choose2}\right)}{6^5}\approx .12345 $$
