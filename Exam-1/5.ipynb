```r
library(reshape2); 
library(ggplot2); 
set.seed(09)
n = 1000; 
x = rnorm(n, 10, 3);
y = x - mean(x); 
z = y/sd(x); 
v = x/sd(x)


sd(v)
mean(x)/sd(x)

xh = hist(x, plot = FALSE); 
yh = hist(y, plot = FALSE); 
zh = hist(z, plot =FALSE); 
vh = hist(v, plot =FALSE); 
plot(xh, col='lightblue', xlim = c(-10,20))
plot(yh, col = 'lightgreen', add = TRUE)
plot(zh, col = 'lightpink', add = TRUE)
plot(vh, col = 'lightslateblue', add = TRUE)
plot(vh, col = 'red')
```
