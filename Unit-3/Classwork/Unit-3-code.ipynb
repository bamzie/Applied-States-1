## Problem #1

**1.(a) Write a function to calculate the expected value of a binomial random variable, using the definition of expectation.**



```r
e = function(n,p){
    x = 0:n
    e = sum(x*dbinom(x,n,p))
    return (e)
}
e(100,1/4)
```

## Problem #2 

In class, we calculated some integrals analytically. Let's do them numerically here.

**2.(a) First, write a function to calculate the expected value of a exponential random variable with lambda = 1/4.**

```r
integrand = function(x, lambda){
    integrand = x*dexp(x,lambda)
    return(integrand)
}
e = function(l){
    e = integrate(integrand, lower = 0, upper = Inf, lambda = l)
    return (e)
}
e(4)
```

We also worked with the pdf $f(x) = \frac{3}{4}(1-x^2)$ on $-1\le x \le 1$ (zero elsewhere).

**2.(b) Find $E(X^3)$.**

```r
n = 100; x = seq(-1,1,length.out = 100)

f = function(x){
    f = x^3*3/4*(1-x^2)
    return(f)
}
plot(x,f(x), type = 'l')
abline(0,0)

integrate(f, 0,1)
integrate(f,-1,0)
```

**2.(c) Estimate $E(X^3)$ using Monte Carlo integration.**

```r
m = 5000; y = runif(m,0,1); x = runif(m,0,1); 
above = sum(y < f(x))/m

y = runif(m,-1,0); x = runif(m,-1,0); 
below = sum(y > f(x))/m
above-below
```
