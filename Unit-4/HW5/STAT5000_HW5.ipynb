## A.1 [6 points] Expectations do not necessarily exist!
Unfortunately, the expected value of a random variable may not be defined. For example, if $\theta$ is an angle uniformly distributed in the interval $(-\pi/2,\pi/2)$, then
$X=\alpha\tan(\theta)$ has the following distribution:

$$ f(x)=\frac{\alpha}{\pi}\frac{1}{x^2+\alpha^2} $$

Show that $E(X)$ is undefined (HINT: compute $E(X)$ using a u-substitution and show that the integral does not converge).

To find $E(X)$ note that,
$$\begin{gather*}
X = \alpha tan(\frac{\pi}{2}) = \infty \\
X = \alpha tan(-\frac{\pi}{2}) = -\infty
\end{gather*}
$$
which will be our bounds for $E(X)$ so, 
$$ \begin{gather*}
\int^{\infty}_{-\infty} x \bigg[\frac{\alpha}{\pi} (\frac{1}{x^2 + \alpha^2})\bigg]dx \\
\frac{2\alpha}{\pi}\int^{\infty}_{-\infty} \frac{x}{x^2 + \alpha^2}dx
\end{gather*}
$$
Let $u = x^2 + \alpha^2$ therefore $du = 2x dx$. Hence: 

$$\begin{gather*}
\frac{\alpha}{\pi}\int^{\infty}_{-\infty} \frac{1}{u}du \\
\frac{\alpha}{\pi}\bigg[\ln{\big(x^2 + \alpha^2\big)}\bigg]^{\infty}_{-\infty} \\
\end{gather*}$$

So if we let $a \rightarrow \infty$ and $b \rightarrow \infty$ we can take the limit of,
$$
\frac{\alpha}{\pi}\bigg[\ln{\big(x^2 + \alpha^2\big)}\bigg]^{a \rightarrow \infty}_{b \rightarrow -\infty} 
$$
Hence, 
$$
\begin{gather*}
\lim_{a\rightarrow \infty} \ln(a^2 + \alpha^2) = \infty \\
\lim_{b\rightarrow -\infty} \ln(b^2 + \alpha^2) = \infty 
\end{gather*}
$$
Ergo, 
$$\begin{gather*}
\frac{\alpha}{\pi}\bigg[\ln{\big(x^2 + \alpha^2\big)}\bigg]^{a \rightarrow \infty}_{b \rightarrow -\infty} = \infty
\end{gather*}$$

## A.3 [4 points] Covariance and independence

**Suppose that X and Y are independent random variables. Prove that $E(XY) = E(X)E(Y)$.**

For a continuous r.v., $X$ and $Y$: 
$$E(XY) = E\big(g(x)h(y)\big) = \int^{\infty}_{-\infty}\int^{\infty}_{-\infty}g(x)h(y)f_x(x)f_y(y)dxdy \newline
\Rightarrow \bigg(\int^{\infty}_{-\infty}g(x)f_x(x)dx\bigg)\bigg(\int^{\infty}_{-\infty}h(y)f_y(y)dy\bigg)\newline
\Rightarrow E\big(g(x)\big)E\big(h(y)\big) \newline
\Rightarrow E(X)E(Y)$$

For a discrete r.v., $X$ and $Y$: 
$$\begin{gather*}
    E(XY) = E\big(g(x)h(y)\big) \\
     = \sum_x\sum_y g(x)h(y)P(X)P(Y) \\
     = \bigg(\sum_x g(x)P(X)\bigg)\bigg(\sum_yh(y)P(Y)\bigg) \\
     = E\big(g(x)\big)E\big(h(y)\big) \\
     = E(X)E(Y)
\end{gather*}$$

**A.4(a) [4 points] Show (without integration), that $sd(\bar{X}) = \frac{\sigma}{\sqrt{n}}$**


$$
\begin{gather*}
Var(\bar{X}) = Var\bigg(\frac{1}{n}\sum^n_{i=1}X_i\bigg) \\
Var(\bar{X})= \frac{1}{n^2}Var\bigg(sum^n_{i=1} X_i \bigg) \\
Var(\bar{X}) = \frac{1}{n^2}\sum^n_{i=1} Var\big(X_i\big) \\
Var(\bar{X}) = \frac{1}{n^2}\cdot \sigma^2n \\
Var(\bar{X}) = \frac{\sigma^2}{n}
\end{gather*} 
$$
And $sd(\bar{X}) = \sqrt{Var(\bar{X})}$: 
$$sd(\bar{X}) = \sqrt{Var(\bar{X})} = \sqrt{\frac{\sigma^2}{n}} = \frac{\sigma}{\sqrt{n}}$$

**A.4(b) [5 points] Show that $\displaystyle\bar{X} = \mu + \frac{\sigma}{\sqrt{n}}Z$ where $Z \sim N(0,1)$.**

We know $Z\sim N(0,1)$ which also mean that $Z = \frac{\bar{X} - \mu}{sd(\bar{X})}$ so: 
$$\begin{gather*}
    Z = \frac{\bar{X} - \mu}{sd(\bar{X})} \\
    Z = \frac{\bar{X} - \mu}{\frac{\sigma}{\sqrt{n}}} \\
    {\frac{\sigma}{\sqrt{n}}}Z = \bar{X} - \mu \\
    \bar{X} = \mu + {\frac{\sigma}{\sqrt{n}}}Z
\end{gather*}$$
