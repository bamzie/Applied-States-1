## Problem 1

**1.(a) In a matrix `x`, generate 500 samples, each of size $n = 50$, from your favorite (non-normal) distribution.**

*Note! In any practical scenario, we would only have one sample of size $n$. The purpose here is the study the distribution of the sample mean, $\bar{x}$, over many (here, $500$) samples. This is an extremely important concept for statistical inference. The distribution of the sample mean, as well as the distribution of other estimators, over many samples is heavily used in the statistical inference portion of the class.*

```r
set.seed(1128)
n =50; m = 500; p = 0.5; 
x = replicate(m, rbinom(n,n,p))
dim(x)
```

**1.(b) Calculate the mean of each sample. This should be a vector of size 500. Call this vector `xbar`.**

`xbar` represents having the mean of $500$ different samples (again: in the real world, we'd only have one of these samples; but this allows us to think about the distribution of the estimator `xbar`).


```r
xbar = colMeans(x)
length(xbar)
```

**1.(c) Construct a histogram of the means. Characterize the distribution of `xbar`. What is the sample mean of `xbar`? The variance of `xbar`?**

```r
x = seq(23,27, length.out = 1000); 
hist(xbar, freq = FALSE, breaks= 15, ylim=c(0,0.9)); 
lines(x, dnorm(x, mean = n*p, sd = sqrt(p*(1-p))))

mean(xbar)
var(xbar)

theorymean = n*p; 
print(theorymean)
theoryvar = p*(1-p); 
print(theoryvar)
```

**1.(d) What theoretical result from class is this a demonstration of?**

Central Limit Theorem

## Problem 2

Suppose that we are going to collect data containing information for $480$ purchases at Monk's Cafe. Further, suppse we know two out of every three purchases at Monk's Cafe are paid for by credit cards.

**2.(a) What is the probability that greater than 300 (of the 480) purchases will be made by credit card?**

```r
n = 480; p =2/3; q = 1-p; 
1 - pbinom(300, n, p); #1 - [F(300)]
```

**2.(b) Now, consider $X \sim N(np, npq)$. Find the probability that $X$ is greater than 300. Compare this to the answer to part 2.(a).**

```r
1 - pnorm(300, n*p, sqrt(n*p*q))
```

**2.(c) Plot the pmf of the binomial random variable used in 2.(a) and the pdf of $X$ from part 2.(b). Comment on the plots.**


```r
x = seq(280, 360); yb = dbinom(x, n, p); yn = dnorm(x, n*p, sqrt(n*p*q))

par(mfrow = c(2,1))
plot(x, yb, type = 'l', xlab = "Binomial Probabilities")
plot(x, yn , type = 'l', xlab = "Normal Probabilities")
```

They both have the same distribution

**2.(d) Find the probability that between $300$ to $320$ will pay for their purchases using credit card. Use both the binomial distribution and the normal approximation given above.**

```r
b = pbinom(320, n,p) - pbinom(300, n, p)
b
normal = pnorm(320, n*p, sqrt(n*p*q)) - pnorm(300, n*p, sqrt(n*p*q))
normal
```

## Problem 3

The amount of impurity in a batch of a chemical product is a random variable with mean value $4.0$ g and standard deviation $1.5$ g. The distribution of this random variable is unknown.

**If 50 batches are independently prepared, what is the (approximate) probability that the average amount of impurity in these $50$ batches is between $3.5$ and $3.8$ g?**

```r
n = 50; xbar_mu = 4; xbar_sig = 1.5/sqrt(n); 
pnorm(3.8, xbar_mu, xbar_sig) - pnorm(3.5, xbar_mu, xbar_sig)
```
