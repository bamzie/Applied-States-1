# Problem 1

**1.(a) Use `rbinom()` to generate $m = 500$ random numbers from a binomial distribution with $n = 50$ and $p = 0.3$. Store these values in `x`. Describe what $n$ and $p$ are, and state any assumptions needed for the binomial distribution to be an appropriate model.**

```r
m = 500; n = 50; p = 0.3; 
x = rbinom(m, n, p)
x
```

$n$ stands for the number of trials, therefore in our binomial we have 50 trails, and $p$ stands for the probability of success in the each trail. 

**1.(b) Print a density histogram of `x`. Describe why the distribution looks the way that it does. Print the mean of `x`. Is it what you might expect? Print the standard deviation of `x`.**

```r
hist(x, freq =FALSE); xBar = mean(x); s = sd(x); 
xBar
s
```

The reason why the histogram looks the way it does is because most of our numbers generated landed around 15 - our mean. The standard deviation represents on average how far away the rest of the generated numbers landed away from our mean.

**3.(a) Generate $m = 500$ random numbers from the geometric distribution where $p = 0.5$. Store these values in `x`.
```r
x = rgeom(500,0.5)
hist(x, prob = TRUE)
```

The geometric distribution represents the number of failures in a sequence of Bernoulli trials before success occurs. With a probability of success $p=0.5$, the probability that we get many failures (e.g., $>6$) will be relatively low.

**3.(b) Note that we can work with the Poisson distribution using `rpois()`. Generate a matrix with $m = 100$ rows and $n = 5$ columns. In the first column, generate $m$ random numbers from the Poisson with $\lambda = 5$. In the second column, generate $m$ random numbers from the Poisson with $\lambda = 10$, and so on, until the last column where $\lambda = 25$. Plot each column. How does the distribution change as $\lambda$ changes?**

```r
m <- 100
trix = cbind(rpois(m, 5), rpois(m, 10), rpois(m, 15), rpois(m,20), rpois(m,25)); 
df = data.frame(trix); 
names(df) = c('lambda = 5', 'lambda = 10', "lambda = 15", 'lambda = 20', 'lambda = 25'); 

library(reshape2); 
library(ggplot2)
ggplot(data = melt(df), mapping = aes(x = value)) + geom_histogram(bins = 20, aes(y=0.5*..density..)) + 
facet_wrap(~variable) + xlim(c(0,40))
```

As $\lambda$ increases we get more of a bell shape curv, approximating to a normal distribution.
